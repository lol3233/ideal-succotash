```python
prompt_tem = ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert product reviewer"),
        ("human","List the main features of the product {product_name}")
    ]
)

```

```python
def analyse_pros(features):
    pros_template =  ChatPromptTemplate.from_messages(
        [
            ("system","You are an expert product reviewer"),
            (
                "human","Given these features: {features}, list the pros of these features"
            )
        ]
    )
    return pros_template.format_prompt(features=features)
```

```python
def analyse_cons(features):
    cons_template =  ChatPromptTemplate.from_messages(
        [
            ("system","You are an expert product reviewer"),
            (
                "human","Given these features: {features}, list the cons of these features"
            )
        ]
    )
    return cons_template.format_prompt(features=features)
```

```python
def combine_pros_cons(pros,cons):
    return f"Pros:\n{pros}\n\nCons:\n{cons}"
```

```python
pros_branch_chain = (
    RunnableLambda(lambda x: analyse_pros(x)) | hub_llm | StrOutputParser()
)
```

```python
cons_branch_chain = (
    RunnableLambda(lambda x: analyse_cons(x)) | hub_llm | StrOutputParser()
)
```

```python
chain = (
    prompt_tem
    | hub_llm
    | RunnableParallel(branches = {"pros": pros_branch_chain, "cons":cons_branch_chain})
    | RunnableLambda(lambda x: combine_pros_cons(x["branches"]["pros"],x["branches"]["cons"]))

)
```

```python
resu = chain.invoke({"product_name" :"Macbook Pro"})
```

```python
print(resu)
```

```python
import os
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import Chroma
from sentence_transformers import SentenceTransformer

# Initialize HuggingFaceHub for other model interactions
hub_llm = HuggingFaceHub(repo_id="mistralai/Mistral-7B-Instruct-v0.3", model_kwargs={"max_new_tokens": 300})

# Load a pre-trained model from Hugging Face for embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = ["This is an example sentence", "Each sentence is converted"]

# Generate embeddings
embeddings = model.encode(sentences)

print(embeddings)
```

```python
file_path = '/workspaces/codespaces-jupyter/latest_news.txt'

# Load the text file
loader = TextLoader(file_path)
documents = loader.load()

# Split the text into smaller chunks
splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=20)
texts = splitter.split_documents(documents)

# Extract text content from Document objects
text_contents = [doc.page_content for doc in texts]

# Generate embeddings for each chunk
embeddings = model.encode(text_contents)
```

```python
persistent_directory = '/workspaces/codespaces-jupyter/'

# Initialize Chroma DB with the persistent directory
chroma_db = Chroma.from_documents(texts, embeddings ,persistent_directory=persistent_directory)

# Add embeddings to Chroma DB
for text, embedding in zip(text_contents, embeddings):
    chroma_db.add(text, embedding)

print("Embeddings have been successfully added to Chroma DB.")
```

```python
import sqlite3
import numpy as np

# Connect to SQLite database (it will create the database file if it doesn't exist)
conn = sqlite3.connect('embeddings.db')

# Create a cursor object to interact with the database
cursor = conn.cursor()
```

```python
# Create a table to store embeddings
cursor.execute('''
CREATE TABLE IF NOT EXISTS embeddings (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    text TEXT,
    embedding BLOB
)
''')
conn.commit()
```

```python
# Insert data into the table
for text, embedding in zip(text_contents, embeddings):
    cursor.execute('''
    INSERT INTO embeddings (text, embedding)
    VALUES (?, ?)
    ''', (text, embedding.tobytes()))
conn.commit()
```

```python
# Query the database to retrieve embeddings
cursor.execute('SELECT * FROM embeddings')
rows = cursor.fetchall()

for row in rows:
    text = row[1]
    embedding = np.frombuffer(row[2], dtype=np.float32)
    print(f"Text: {text}, Embedding: {embedding}")
```

```python
# Close the connection
conn.close()
```

```python
def search(query, model, k=3):
    # Generate embedding for the query
    query_embedding = model.encode([query])[0]
    
    # Retrieve all embeddings from the database
    cursor.execute('SELECT text, embedding FROM embeddings')
    rows = cursor.fetchall()
    
    # Calculate cosine similarity between query embedding and stored embeddings
    similarities = []
    for row in rows:
        text = row[0]
        embedding = np.frombuffer(row[1], dtype=np.float32)
        similarity = np.dot(query_embedding, embedding) / (np.linalg.norm(query_embedding) * np.linalg.norm(embedding))
        similarities.append((similarity, text))
    
    # Sort by similarity and return top k results
    similarities.sort(reverse=True, key=lambda x: x[0])
    results = [text for _, text in similarities[:k]]
    return results

# Example usage
query = "latest news on fire"
results = search(query, model)
for result in results:
    print(result)
```
